```python
from typing import Callable, Tuple
from finite_distributions.FiniteDistribution import FiniteDistribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sinkhorn.SinkhornRunner as SinkhornRunner
import sinkhorn.SinkhornKernels as skern
import visualizer.joint_distribution_visualizer as jdv
```

```python
coin_1 = FiniteDistribution({"H1": 0.5, "T1": 0.5})
coin_2 = FiniteDistribution({"H2": 0.5, "T2": 0.5})
```

```python
costFromGamma = lambda gam: lambda x, y: (0 if x[0] == y[0] else gam)
```

```python
gamma = 1.

# p-norm
p = 2.
sinkhorn_runner = skern.get_pnorm_regularized_runner(p, costFromGamma(gamma))
# entropic
# sinkhorn_runner = skern.get_entropically_regularized_runner(costFromGamma(gamma))


epsilon = 1.
delta = 0.0001

pi, f, g = sinkhorn_runner.run_sinkhorn(coin_1, coin_2, epsilon, delta)

jdv.visualize_joint_probability(pi, title = "Joint Probability", annot = True)

product = coin_1.productDistribution(coin_2)
jdv.visualize_joint_probability(product, title = "Product Probability", annot = True)
```
